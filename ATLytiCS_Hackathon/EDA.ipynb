```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder


import warnings
# Suppress annoying harmless error.
warnings.simplefilter('ignore')
```

```python
# Load survey data into dataframe

df=pd.read_pickle("./demograph.pkl")
```

```python
labels = ['Bought BEV', 'Did Not Buy BEV']
index = np.arange(len(labels))
ax = sns.countplot(x="BEV", data=df)
plt.xticks(index, labels)
plt.show()
```

```python
consumers = df[df['BEV']==1]
consumers['Gender'] = np.where(consumers['Sex']==1, 'Male','Female') 
#df['Bought'] = np.where(df['BEV'] == 0, 'Did Not Buy BEV', 'Bought BEV')
index = np.arange(len(labels))
ax = sns.countplot(x="Gender",  data=consumers)
plt.ylabel('Number of Respondents',fontsize=16, labelpad=10)
plt.xticks(index,fontsize=14,rotation = 0)
plt.yticks(fontsize=14,rotation = 0)
plt.show()
```

```python
consumers.columns
```

```python
consumers['Activity']= consumers['Primary Activity'].replace({1:'Working',
                                                              2:'Temporarily absent from a job',
                                                              3:'Looking for work / unemployed',
                                                              4:'A homemaker',
                                                              5:'Going to school',
                                                              6:'Retired'}).copy()

labels = ['Working','Temporarily absent from a job', 'Looking for work / unemployed',
         'A homemaker','Going to school','Retired'] 
index = np.arange(len(labels))
ax = sns.countplot(x="Activity",  data=consumers)
plt.xticks(index, labels,fontsize=14,rotation = 90)

plt.show()
```

```python
consumers['Life Cycle Label']=consumers['Life Cycle'].replace({1:'one adult, no children',   
                                                              2:'2+ adults, no children',
                                                              3:'one adult, youngest child 0-5',
                                                              4:'2+ adults, youngest child 0-5',
                                                              5:'one adult, youngest child 6-15',
                                                              6: '2+ adults, youngest child 6-15',
                                                              7:'one adult, youngest child 16-21',
                                                              8: '2+ adults, youngest child 16-21',
                                                              9: 'one adult, retired, no children',
                                                              10: '2+ adults, retired, no children'}).copy()
    
ax = sns.countplot(x="Life Cycle Label",  data=consumers)
plt.xticks(fontsize=14,rotation = 90)

plt.show()






```

```python
ax = sns.countplot(x="Family Income",  data=consumers)
plt.xticks(fontsize=14,rotation = 90)

plt.show()
```

```python
ax = sns.distplot(consumers['Age'])
plt.xticks(fontsize=14,rotation = 45)

plt.show()
```

```python
labels = ['Yes', 'No', 'Other Arrangement'] 
index = np.arange(len(labels))

ax = sns.countplot(x='Home Ownership',  data=consumers)
plt.xticks(index, labels,fontsize=14,rotation = 90)

plt.show()
```
